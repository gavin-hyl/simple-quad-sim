```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def extract(path=None):
    with open(path, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='|')
        x = []
        y = []
        z = []
        for row in reader:
            x.append(float(row[0]))
            y.append(float(row[1]))
            z.append(float(row[2]))
    return x, y, z

x, y, z = extract('2a-wind.csv')
plt.plot(x, 'r', label='F_0=0.1')

x1, y1, z1 = extract('2a-wind2.csv')
plt.plot(x1,'b', label='F_0=1')

x2, y2, z2 = extract('2a-wind3.csv')
plt.plot(x2,'g', label='F_0=0.5')

plt.xlabel('Time step')
plt.ylabel('x_pos (desired = 0)')

plt.legend()
```

```python
plt.plot(z, 'r', label='F_0=0.1')
plt.plot(z1,'b', label='F_0=1')
plt.plot(z2,'g', label='F_0=0.5')
```
