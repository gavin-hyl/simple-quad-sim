```python
import csv
import matplotlib.pyplot as plt
```

```python
TRANSIENT_TICS = 750
def extract(path=None):
    with open(path, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='|')
        x = []
        y = []
        z = []
        tic = 0
        for row in reader:
            tic += 1
            if tic < TRANSIENT_TICS:
                continue
            x.append(float(row[0]))
            y.append(float(row[1]))
            z.append(float(row[2]))
    return x, y, z

wind_speeds = [0, 1, 3, 5, 10]
for i, wind_speed in enumerate(wind_speeds):
    x, y, z = extract(f'{wind_speeds[i]}_sin_wind-hover.csv')
    plt.plot(x, label=f'F_0={wind_speeds[i]}')

plt.xlabel(f'Time step (offset by {TRANSIENT_TICS} steps)')
plt.ylabel('x_pos (desired = 0)')

plt.legend()
```
